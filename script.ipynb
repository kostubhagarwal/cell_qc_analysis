```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "data.xlsx"

def import_data(file):
    data = pd.read_excel(file)
    return data

def plot(data, name):
    index = data.index.tolist()
    plt.scatter(index, data)
    plt.xlabel("index")
    plt.ylabel(name)
    plt.title(name + " distribution")
    plt.show()

def remove_outliers(data, threshold):
    mean = data.mean()
    std = data.std()
    z_score = (data - mean) / std
    outliers = data[abs(z_score) > threshold]
    return outliers
```

```python
import_data(file)

ir_data = data.iloc[:, 1]
ocv_data = data.iloc[:, 2]
```

```python
plot(ir_data, "IR")

cleaned_data = remove_outliers(ir_data,3)
print("IR data after cleaning:", cleaned_data)
```

```python
plot(ocv_data, "OCV")

cleaned_data = remove_outliers(ocv_data,3)
print("OCV data after cleaning:", cleaned_data)
```
